```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
sns.set(rc={'figure.figsize':(20,10)})
```

```python
df_energy = pd.read_csv("drive/MyDrive/Trab/energy.csv", sep=';')
df_energy = df_energy[['country', 'year', 'hydro_electricity', 'wind_electricity']]
df_energy = df_energy.drop(df_energy[df_energy.year < 2010].index)
df_energy = df_energy.drop(df_energy[df_energy.year > 2020].index)

df_energy
```

```python
df_mask = df_energy['country'] == 'Bolivia'
df_mask += df_energy['country'] == 'Peru'
filtered_df = df_energy[df_mask]

filtered_df = filtered_df.set_index("year")
filtered_df = filtered_df.sort_index()
filtered_df = filtered_df.pivot(columns='country')
filtered_df['hydro_electricity'] = filtered_df['hydro_electricity'].astype(float)
filtered_df['wind_electricity'] = filtered_df['wind_electricity'].astype(float)

filtered_df.corr(method='pearson')
```
